```python
from SNN import SNN, chain, fully_connect, connect
import matplotlib.pyplot as plt

net = SNN(N=3, T=2000, V_rest=0, V_thresh=15, tau=0.9, ref_period=20, A_plus=20, A_minus=20, tau_stdp=20, w_init=1.0)
chain(net, [0, 1, 2])

for cycle in range(net.T // 60):
    base = cycle * 60 + 1
    if base < net.T: net.stim[0][base] = 20
    if base + 20 < net.T: net.stim[1][base + 20] = 20
    if base + 40 < net.T: net.stim[2][base + 40] = 20

net.run()
print([t % 60 for t in net.spike_times[1][:30]])
print([t % 60 for t in net.spike_times[2][:30]])
net.plot()
```
